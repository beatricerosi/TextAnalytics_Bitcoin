```python
import numpy as np 
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('...\groupBy.csv', delimiter=',', lineterminator='\n', low_memory=False)
dataset
```

```python
# for splitting dataset into train set and test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(dataset["tweet"],dataset["value\r"], test_size = 0.33, random_state = 10)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# for vectorizing words
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vect = CountVectorizer(stop_words='english')
vect.fit(X_train)
```

```python
print(vect.get_feature_names()[0:20])
print(vect.get_feature_names()[-20:])
```

```python
X_train_df = vect.transform(X_train)
X_test_df = vect.transform(X_test)
type(X_test_df)
```

```python
prediction = dict()
# Naive Bayes Machine Learning Model
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
model = BernoulliNB()
model.fit(X_train_df,y_train)
```

```python
#faccio la predizione su X_test usando il naive bayes
prediction["naive_bayes"] = model.predict(X_test_df)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# accuracy
accuracy_score(y_test,prediction["naive_bayes"])
```

```python
#classification report
print(classification_report(y_test, prediction['naive_bayes'], target_names = ["bear", "bull"]))
```
